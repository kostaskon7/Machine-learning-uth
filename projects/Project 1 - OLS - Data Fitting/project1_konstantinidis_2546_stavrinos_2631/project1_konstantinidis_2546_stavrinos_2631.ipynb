#### Problem 2.1

> Derive this last equation from the system
> $$A\mathbf{c} = \mathbf{a}$$
> spelled out for this special case.

---
#### Solution 2.1

Given the initial equations
$$ c_0 + c_1x = y $$
$$A\mathbf{c} = 
\left[\begin{array}{cc}1 & x_0\\1 & x_1\end{array}\right]
\left[\begin{array}{c}c_0\\c_1\end{array}\right]
=\
\left[\begin{array}{c}y_0\\y_11\end{array}\right]
= \mathbf{a}$$

Using Gaussian elimination, 

$$\left[\begin{array}{cc|c}1 & x_0 & y_0\\1 & x_1 & y_1\end{array}\right]
\rightarrow
\left[\begin{array}{cc|c}1 & x_0 & y_0\\0 & x_1 - x_0 & y_1-y_0\end{array}\right]$$

This gives us the following two equations: 

$$c_0 + c_1x_0 = y_0$$
$$ c_0(x_1-x_0) = y_1 - y_0$$

Rearranging the equations for $c_0$ and $c_1$: 

$$ c_0 = y_0 - c_1x_0$$ 
$$ c_1 = \frac{y_1 - y_0}{x_1-x_0}$$

Substituting the second equation into the first equation: 

$$ c_0 = y_0 - \frac{y_1 - y_0}{x_1-x_0}x_0$$ 

Substituting $c_0$ and $c_1$ into the first equation: 

$$y_0 - \frac{y_1 - y_0}{x_1-x_0}x_0 + \frac{y_1 - y_0}{x_1-x_0}x = y$$

Rearranging, 

$$ y - y_0 = \frac{y_1 - y_0}{x_1 - x_0}\ (x - x_0) $$

#### Problem 3.2

> Find expressions for $c_0$ and $c_1$ in terms of $N, X, Y, S, P$ by solving the normal equations. Recall that
> $$\left[\begin{array}{cc}a & b\\c & d \end{array}\right]^{-1} \;=\; 
\frac{1}{ad-bc}\ \left[\begin{array}{rr}d & -b\\-c & a \end{array}\right]$$

---
#### Solution 3.2

First, we solve for $\mathbf{a}$, 

$$ A^T A \mathbf{c} = A^T \mathbf{a}$$

Multiplying both sides by the inverse of $A^T A$
$$ (A^T A)^{-1} (A^T A) \mathbf{c} = (A^T A)^{-1} A^T \mathbf{a} $$
$$\mathbf{c} = (A^T A)^{-1} A^T \mathbf{a} $$

Using the definition of the the inverse of a 2x2 matrix, 

$$\left[\begin{array}{cc}a & b\\c & d \end{array}\right]^{-1} \;=\; 
\frac{1}{ad-bc}\ \left[\begin{array}{rr}d & -b\\-c & a \end{array}\right]$$

$$ (A^T A)^{-1} 
=\ 
\left[\begin{array}{cc}N & X\\ X & S\end{array}\right]^{-1}
=\ 
\frac{1}{NS - X^2} 
\left[\begin{array}{cc}S & -X\\ -X & N\end{array}\right]
$$

Substituting into the equation above, 

$$\mathbf{c} 
=\
\left[\begin{array}{c}c_0\\ c_1\end{array}\right]
=\
\frac{1}{NS - X^2} 
\left[\begin{array}{cc}S & -X\\ -X & N\end{array}\right]
\left[\begin{array}{c}Y\\P\end{array}\right]
=\
\frac{1}{NS - X^2} 
\left[\begin{array}{c}SY - XP\\ -XY + NP\end{array}\right]
$$

#### Solutions 4.1

```python
def fit(x,y,k):
    '''
    This function takes `numpy` vectors `x` and `y` with $N$ coordinates each 
    and a degree `k` (a nonnegative integer), and returns a `numpy` vector `c` 
    with the coefficients of the polynomial of degree `k` that fits 
    the points `(x[n], y[n])` best in the Least-Squares sense.
    '''

    N = len(y)
    arr = np.zeros((N, k+1))
    for i in range(N):
        for j in range(k+1):
            arr[i, j] = x.item(i) ** j
        
    A = np.matrix(arr)
    a = np.reshape(y, [y.size, 1])
    
    c = np.linalg.lstsq(A, a, rcond=None)[0]
    return c
```

## LU Decomposition

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from scipy.linalg import lu

def fit(x,y,k):
    N = len(y)
    
    A = np.array([[x.item(i) ** j for j in range(k+1)] for i in range(N)])
    
    #Forms the normal equations
    a=(A.T).dot(A)
    b=(A.T).dot(y)
   
    # calculate coefficients
    P,L,U = lu(a)
    c = inv(P.dot(L.dot(U))).dot(b)
   
    print("\nThe coefficients for k =",k,"are :\n",c)
    # predict using coefficients
    yhat = A.dot(c.T)
    
    # calculate the error
    e=( sum((yhat-y)**2 )/N )**0.5
    print("\nThe error for k =",k,"is :",e)
    
    #plot data and predictions
    plt.scatter(x, y)
    plt.plot(x, yhat, color='red')
    plt.show()
    return [yhat,e]

x = np.linspace(-5,5,60)
y = 1/(1+x**2)

[y_3,e_3]=fit(x,y,3)

[y_7,e_7]=fit(x,y,7)

[y_15,e_15]=fit(x,y,15)

[y_21,e_21]=fit(x,y,21)


print("\nThe polynomials in the same graph\n")
plt.scatter(x, y)
plt.plot(x,y_3,x, y_7,x,y_15,x,y_21)

plt.show()
print("\nThe erros in the same graph\n")

e_all=[e_3, e_7,e_15,e_21]
zeros=[0,0,0,0]
labels=["n3","n7","n15","n21"]

plt.plot(e_all,zeros,'bo-')

for i in range(0,4):

    label = "{:.2f}".format(zeros[i])

    plt.annotate(labels[i], # this is the text
                 (e_all[i],zeros[i]), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
```

## LU Decomposition

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from scipy.linalg import lu
import math

def fit(x,y,k):
    N = len(y)
    
    w = (2*math.pi)/24;
    
    A = np.zeros((N, k+1))
    for i in range(N):
        temp=0
        for j in range(k+1):
            if(j % 2 != 0):
                A[i, j] = math.sin(x.item(i) * temp* w)
            else:
                A[i, j] = math.cos(x.item(i) *temp * w)
                temp= temp+1
    
    #Forms the normal equations
    a=(A.T).dot(A)
    b=(A.T).dot(y)
   
    # calculate coefficients
    P,L,U = lu(a)
    c = inv(P.dot(L.dot(U))).dot(b)
   
    print("\nThe coefficients for k =",k,"are :\n",c)
    # predict using coefficients
    yhat = A.dot(c.T)
    
    # calculate the error
    e=( sum((yhat-y)**2 )/N )**0.5
    print("\nThe error for k =",k,"is :",e)
    
    #plot data and predictions
    plt.scatter(x, y)
    plt.plot(x, yhat, color='red')
    plt.show()
    return [yhat,e]

x  = np.arange(0, 25, 1)
y = np.array([110.49, 73.72, 23.39, 17.11, 20.31, 29.37, 74.74117, .02, 298.04, 348.13, 294.75, 253.78, 250.48, 239.48236, .52, 245.04, 286.74, 304.78, 288.76, 247.11, 216.73, 185.78, 171.19, 171.73, 164.05])

[y_3,e_3]=fit(x,y,3)

[y_7,e_7]=fit(x,y,7)

[y_15,e_15]=fit(x,y,15)

[y_21,e_21]=fit(x,y,21)


print("\nThe trigonometric functions in the same graph\n")
plt.scatter(x, y)
plt.plot(x,y_3,x, y_7,x,y_15,x,y_21)

plt.show()
print("\nThe erros in the same graph\n")

e_all=[e_3, e_7,e_15,e_21]
zeros=[0,0,0,0]
labels=["n3","n7","n15","n21"]

plt.plot(e_all,zeros,'bo-')

for i in range(0,4):

    label = "{:.2f}".format(zeros[i])

    plt.annotate(labels[i], # this is the text
                 (e_all[i],zeros[i]), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
```

## QR Decomposition

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from numpy.linalg import qr
import math

def fit(x,y,k):
    N = len(y)
    w = (2*math.pi)/24;
    
    A = np.zeros((N, k+1))
    for i in range(N):
        temp=0
        for j in range(k+1):
            if(j % 2 != 0):
                A[i, j] = math.sin(x.item(i) * temp* w)
            else:
                A[i, j] = math.cos(x.item(i) *temp * w)
                temp= temp+1
    

    #Forms the normal equations
    a=(A.T).dot(A)
    b=(A.T).dot(y)
    # calculate coefficients
    Q, R = qr(a)
    c = inv(R).dot(Q.T).dot(b)
    print("\nThe coefficients for k =",k,"are :\n",c)
    # predict using coefficients
    yhat = A.dot(c.T)
    
    
    # calculate the error
    e=( sum((yhat-y)**2 )/N )**0.5
    print("\nThe error for k =",k,"is :",e)
    
    #plot data and predictions
    plt.scatter(x, y)
    plt.plot(x, yhat, color='red')
    plt.show()
    return [yhat,e]

x  = np.arange(0, 25, 1)
y = np.array([110.49, 73.72, 23.39, 17.11, 20.31, 29.37, 74.74117, .02, 298.04, 348.13, 294.75, 253.78, 250.48, 239.48236, .52, 245.04, 286.74, 304.78, 288.76, 247.11, 216.73, 185.78, 171.19, 171.73, 164.05])

[y_3,e_3]=fit(x,y,3)

[y_7,e_7]=fit(x,y,7)

[y_15,e_15]=fit(x,y,15)

[y_21,e_21]=fit(x,y,21)


print("\nThe trigonometric functions in the same graph\n")
plt.scatter(x, y)
plt.plot(x,y_3,x, y_7, x,y_15,x,y_21)
plt.show()
print("\nThe erros in the same graph\n")

e_all=[e_3, e_7,e_15,e_21]
zeros=[0,0,0,0]
labels=["n3","n7","n15","n21"]

plt.plot(e_all,zeros,'bo-')

for i in range(0,4):

    label = "{:.2f}".format(zeros[i])

    plt.annotate(labels[i], # this is the text
                 (e_all[i],zeros[i]), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
```

## SVD Decomposition

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import pinv
import math

def fit(x,y,k):
    N = len(y)
    
    w = (2*math.pi)/24;
    
    A = np.zeros((N, k+1))
    for i in range(N):
        temp=0
        for j in range(k+1):
            if(j % 2 != 0):
                A[i, j] = math.sin(x.item(i) * temp* w)
            else:
                A[i, j] = math.cos(x.item(i) *temp * w)
                temp= temp+1
    
    a=(A.T).dot(A)
    b=(A.T).dot(y)
   
    # calculate coefficients
    c = pinv(a).dot(b)
    print("\nThe coefficients for k =",k,"are :\n",c)
    # predict using coefficients
    yhat = A.dot(c.T)
    
    # calculate the error    
    e=( sum((yhat-y)**2 )/N )**0.5
    print("\nThe error for k =",k,"is :",e)
    
    # plot data and predictions
    plt.scatter(x, y)
    plt.plot(x, yhat, color='red')
    plt.show()
    return [yhat,e]


x  = np.arange(0, 25, 1)
y = np.array([110.49, 73.72, 23.39, 17.11, 20.31, 29.37, 74.74117, .02, 298.04, 348.13, 294.75, 253.78, 250.48, 239.48236, .52, 245.04, 286.74, 304.78, 288.76, 247.11, 216.73, 185.78, 171.19, 171.73, 164.05])


[y_3,e_3]=fit(x,y,3)

[y_7,e_7]=fit(x,y,7)

[y_15,e_15]=fit(x,y,15)

[y_21,e_21]=fit(x,y,21)


print("\nThe trigonometric functions in the same graph\n")
plt.scatter(x, y)
plt.plot(x,y_3,x, y_7,x,y_15, x,y_21)
plt.show()
print("\nThe erros in the same graph\n")

e_all=[e_3, e_7,e_15,e_21]
zeros=[0,0,0,0]
labels=["n3","n7","n15","n21"]

plt.plot(e_all,zeros,'bo-')

for i in range(0,4):

    label = "{:.2f}".format(zeros[i])

    plt.annotate(labels[i], # this is the text
                 (e_all[i],zeros[i]), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
```
